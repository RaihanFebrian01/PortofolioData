```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
#%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/MyDrive/kuliah/SEMESTER 5/Data Mining/TUBES/Terbaru/fix/kamis26.csv'
df = pd.read_csv(path)
df.head
```

```python
 k_rng = range(1,10)
 sse = []
 for k in k_rng:
   km=KMeans(n_clusters=k)
   km.fit(df[['NilaiAkreditasi','JumlahSiswa']])
   sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
plt.scatter(df['NilaiAkreditasi'],df['JumlahSiswa'])
plt.xlabel('Nilai')
plt.ylabel('jumlah siswa')
```

```python
km = KMeans(n_clusters=3, random_state=42)
km
```

```python
y_predicted = km.fit_predict(df[['NilaiAkreditasi','JumlahSiswa']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.NilaiAkreditasi,df1['JumlahSiswa'], color='green')
plt.scatter(df2.NilaiAkreditasi,df2['JumlahSiswa'], color='yellow')
plt.scatter(df3.NilaiAkreditasi,df3['JumlahSiswa'], color='red')
plt.xlabel('Nilai')
plt.ylabel('JumlahSiswa')
plt.legend
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['JumlahSiswa']])
df['JumlahSiswa'] = scaler.transform(df[['JumlahSiswa']])

scaler.fit(df[['NilaiAkreditasi']])
df['NilaiAkreditasi'] = scaler.transform(df[['NilaiAkreditasi']])
df.head
```

```python
km=KMeans(n_clusters=3, random_state=42)
y_predicted=km.fit_predict(df[['NilaiAkreditasi','JumlahSiswa']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.NilaiAkreditasi,df1['JumlahSiswa'], color='green')
plt.scatter(df2.NilaiAkreditasi,df2['JumlahSiswa'], color='yellow')
plt.scatter(df3.NilaiAkreditasi,df3['JumlahSiswa'], color='red')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black',marker='*',label='centroid')
plt.xlabel('Nilai')
plt.ylabel('JumlahSiswa')
plt.legend
```
